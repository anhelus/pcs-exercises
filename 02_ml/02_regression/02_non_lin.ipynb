```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_percentage_error as mape
```

```python
X, y = make_regression(
    n_samples=100,
    n_features=1,
    n_informative=1,
    noise=10,
    random_state=42)

p = X.argsort(axis=0)
X = X[p].reshape(-1, 1)
y = y[p].reshape(-1)
```

```python
lr = LinearRegression()
lr.fit(X, y)
y_pred_lr = lr.predict(X)
lr_mape = mape(y, y_pred_lr)
```

```python
dtr = DecisionTreeRegressor()
dtr.fit(X, y)
y_pred_dtr = dtr.predict(X)
dtr_mape = mape(y, y_pred_dtr)
```

```python
rfr = RandomForestRegressor()
rfr.fit(X, y)
y_pred_rfr = rfr.predict(X)
rfr_mape = mape(y, y_pred_rfr)
```

```python
mlpr = MLPRegressor()
mlpr.fit(X, y)
y_pred_mlpr = mlpr.predict(X)
mlpr_mape = mape(y, y_pred_mlpr)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred_lr, c='g', label=f'Linear Regression - MAPE: {lr_mape}')
plt.plot(X, y_pred_dtr, c='r', label=f'Decision Tree - MAPE: {dtr_mape}')
plt.plot(X, y_pred_rfr, c='y', label=f'Random Forest - MAPE: {rfr_mape}')
plt.plot(X, y_pred_mlpr, c='k', label=f'MLP - MAPE: {mlpr_mape}')
plt.grid()
plt.legend()
plt.show()
```
